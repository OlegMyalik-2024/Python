```python
import random
import datetime
# import prettytable                  # пакет для таблицы
import matplotlib.pyplot as plt     # библиотека для графика
import numpy as np

def BubbleSort(A):                  # сортировка пузырьком
    for i in range(len(A)):
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                a = A[j]
                A[j] = A[j+1]
                A[j+1] = a

def InsertSort(A):                 #сортировка вставками  
    for i in range(1, len(A)):
        t=A[i]
        j=i-1
        while j>=0 and A[j]>t:
            A[j+1]=A[j]
            j-=1
        A[j+1]=t

def ShakerSort(A):                #шейкерная сортировка
    left=0
    right=len(A)-1
    while left<=right:
        for i in range(left, right, +1):
            if A[i]>A[i+1]:
                A[i], A[i+1]=A[i+1], A[i]
        right-=1
        for i in range(right, left, -1):
            if A[i-1]>A[i]:
                A[i], A[i-1]=A[i-1], A[i]
        left+=1



def QuickSort(A, fst, lst):         # быстрая сортировка
    if fst >= lst:
        return
    #i, j = fst, lst
    pivot = A[fst]
    # pivot = A[random.randint(fst, lst)]
    first_bigger = fst+1
    while first_bigger <= lst:
        if A[first_bigger] >= pivot:
            break
        first_bigger += 1
    i = first_bigger+1
    while i <= lst:
        if A[i] < pivot:
            A[i], A[first_bigger] = A[first_bigger], A[i]
            first_bigger += 1
        i += 1
    last_smaller = first_bigger-1
    A[fst], A[last_smaller] = A[last_smaller], A[fst]
    QuickSort(A, fst, last_smaller-1)
    QuickSort(A, first_bigger, lst)




# table = prettytable.PrettyTable(["Размер списка", "Время пузырька", "Время быстрой"])
x=[]
y1=[]
y2=[]
y3=[]
y4=[]




for N in range(1000,5001,1000):
    x.append(N)
    min=1
    max=N
    A=[]
    for i in range (N):
        A.append(int(round(random.random()*(max-min)+min)))

    #print(A)

    B = A.copy()
    C=[]
    С = A.copy()
    D=[]
    D = A.copy()
    # print(B)

    # BubbleSort(A)
    print("---")
    # print(A)


    # InsertSort(C)
    print("---")
    # print(C)


    # ShakerSort(D)
    print("---")
    # print(D)



    # QuickSort(B, 0, len(B)-1)
    print("---")
    # print(B)

    t1 = datetime.datetime.now()
    BubbleSort(A)
    t2 = datetime.datetime.now()
    y1.append((t2-t1).total_seconds())
    print("Пузырьковая сортировка   " +str(N)+"   заняла   "+str((t2-t1).total_seconds()) + "c")


    t5 = datetime.datetime.now()
    InsertSort(C)
    t6 = datetime.datetime.now()
    y3.append((t6-t5).total_seconds())
    print("Сортировка вставками   " +str(N)+"   заняла   "+str((t6-t5).total_seconds()) + "c")


    t7 = datetime.datetime.now()
    InsertSort(D)
    t8 = datetime.datetime.now()
    y4.append((t8-t7).total_seconds())
    print("Шейкерная сортировка   " +str(N)+"   заняла   "+str((t8-t7).total_seconds()) + "c")

    t3 = datetime.datetime.now()
    QuickSort(B, 0, len(B)-1)
    t4 = datetime.datetime.now()
    y2.append((t4 - t3).total_seconds())
    print("Быстрая   " +str(N)+"   заняла   "+str((t4-t3).total_seconds()) + "c")






    # table.add_row([str(N), str((t2-t1).total_seconds()), str((t4-t3).total_seconds())])
# print(table)
#Общий график сортировки данных
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Общий график сортировки данных", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)

plt.plot(x, y1, "C1")
plt.plot(x, y3, "C3")
plt.plot(x, y4, "C4")
plt.plot(x, y2, "C2")

plt.show()

#Пузырьковая сортировка
plt.title("График сортировки данных (пузырьковая сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y1, "C1")
plt.show()

#Сортировка вставками
plt.title("График сортировки данных (сортировка вставками)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y3, "C3")
plt.show()

#Шейкерная сортировка
plt.title("График сортировки данных (шейкерная сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y4, "C4")
plt.show()

#Быстрая сортировка
plt.title("График сортировки данных (быстрая сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y2, "C2")
plt.show()
#Точечный график сортировки данных
#Пузырьковая сортировка
plt.title("Точечный график сортировки данных (пузырьковая сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y1, 'o')
plt.show()

#Сортировка вставками
plt.title("Точечный график сортировки данных (сортировка вставками)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y3, 'o')
plt.show()

#Шейкерная сортировка
plt.title("Точечный график сортировки данных (шейкерная сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y4, 'o')
plt.show()

#Быстрая сортировка
plt.title("Точечный график сортировки данных (быстрая сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
plt.plot(x, y2, 'o')
plt.show()

#BarPlot для сортировок данных
#Пузырьковая сортировка
plt.title("BarPlot график сортировки данных (пузырьковая сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
x = np.array(["1000", "2000", "3000", "4000", "5000"])
plt.bar(x,y1)
plt.show()

#Сортировка вставками
plt.title("BarPlot график сортировки данных (сортировка вставками)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
x = np.array(["1000", "2000", "3000", "4000", "5000"])
plt.bar(x,y3)
plt.show()

#Шейкерная сортировка
plt.title("BarPlot график сортировки данных (шейкерная сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
x = np.array(["1000", "2000", "3000", "4000", "5000"])
plt.bar(x,y4)
plt.show()

#Быстрая сортировка
plt.title("BarPlot график сортировки данных (быстрая сортировка)", fontdict = font1)
plt.xlabel("Количество элементов", fontdict = font2)
plt.ylabel("Время выполнения сортировки (с)", fontdict = font2)
x = np.array(["1000", "2000", "3000", "4000", "5000"])
plt.bar(x,y2)
plt.show()
```
